```python
import numpy as np
import pandas as pd
```

```python
memoria = pd.DataFrame(np.zeros((16,6)), columns = ['e1', 'e0', 'b', 'sp', 'e0\'', 'e1\''])

#Preencher a SP
for i in range(16):
  if (i >> 0) % 2 != 0:
    memoria['sp'].iloc[i] = 1

#Preencher o B
for i in range(16):
  if (i >> 1) % 2 != 0:
    memoria['b'].iloc[i] = 1

#Preencher o E0
for i in range(16):
  if (i >> 2) % 2 != 0:
    memoria['e0'].iloc[i] = 1

#Preencher E1
for i in range(16):
  if (i >> 3) % 2 != 0:
    memoria['e1'].iloc[i] = 1

#Preenchendo E0'
memoria['e0\''].iloc[2] = 1
memoria['e0\''].iloc[4] = 1

#Preenchendo E1'
memoria['e1\''].iloc[5] = 1
memoria
```

```python
def controle(estado, botao, sensor_porta):
  
  #Lógica de transição

  if estado == 0 and botao == 1:
      estado = 1

  elif estado == 1 and sensor_porta == 1:
      estado = 2

  elif estado == 2 and botao == 1:
      estado = 0

#Lógica de ação

  if estado == 0:
    buzina = 0
    luz_vermelha = 0

  if estado == 1:
    buzina = 0
    luz_vermelha = 1

  if estado == 2:
    buzina = 1
    luz_vermelha = 1
    
  if estado == 3:
    buzina = 0
    luz_vermelha = 0

  return estado, buzina, luz_vermelha
```
